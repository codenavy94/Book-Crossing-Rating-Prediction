```python
import pandas as pd
```

```python
all_data = pd.read_csv('data_new/all_data_with_age_group_and_author_cluster.csv')
movies = pd.read_csv('data_new/rated_movie_info.csv')
```

```python
all_data.head()
```

```python
movies.head()
```

```python
movies_rgb = movies.drop(columns=['Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Author_2cluster', 'Author_3cluster'])
movies_rgb.head()
```

```python
total_data = pd.merge(all_data, movies_rgb, how='left', on='ISBN')
```

```python
total_data.head()
total_data.to_csv('data_new/final_data.csv', index=False, encoding='utf-8')
```
