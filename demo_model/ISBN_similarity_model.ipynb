```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
```

```python
all = pd.read_csv('data_new/all_data_final.csv')
all.columns
```

```python
users = pd.read_csv('data_new/users_final.csv')
users = users.set_index('User-ID')
users.head()
```

```python
ratings = all[['User-ID', 'ISBN', 'Age-Group', 'Book-Rating']]
ratings.head()
```

```python
movies = pd.read_csv('data_new/rated_movie_info.csv')
movies.head()
```

```python
movie_scaled = pd.read_csv('data_new/movie_info_scaled.csv')
movie_scaled = movie_scaled.set_index('ISBN')
```

```python
movie_similarity = cosine_similarity(movie_scaled, movie_scaled) # (22089, 22089)
movie_similarity = pd.DataFrame(movie_similarity, index=movie_scaled.index, columns=movie_scaled.index)
movie_similarity.head()
```

```python
# scaled된 movie 정보 불러오기
book_scaled = pd.read_csv('data_new/book_info_scaled.csv')
book_scaled = book_scaled.set_index('ISBN')

# movie간 유사도 구하기
book_similarity = cosine_similarity(book_scaled, book_scaled) # (22089, 22089)
book_similarity = pd.DataFrame(book_similarity, index=book_scaled.index, columns=book_scaled.index)

def get_knn_ISBNs(ISBN, k=10):
    
    knn_result = book_similarity.loc[ISBN].sort_values(ascending=False)
    knn_lst = list(zip(knn_result.index, knn_result.values))[1:k+1]
    return knn_lst
```

```python
train, test, y_train, y_test = train_test_split(ratings, ratings['Book-Rating'], test_size=0.3, random_state=42)
```

```python
# 도서 전체의 전체 평점평균 구하기
total_mean_rating = train['Book-Rating'].mean()
print(total_mean_rating)

# ISBN별(도서별) 전체 평점평균 구하기
mean_rating_per_ISBN = train.groupby('ISBN')['Book-Rating'].mean()

# 사용자 연령별 전체 평점평균 구하기
mean_rating_per_AgeGroup = train.groupby('Age-Group')['Book-Rating'].mean()
print(mean_rating_per_AgeGroup[0])

# ISBN별(도서별) 연령대에 따른 평점평균 구하기
mean_rating_per_age_group_and_ISBN = train.groupby(['ISBN', 'Age-Group'])['Book-Rating'].mean().to_frame().reset_index()
mean_rating_per_age_group_and_ISBN.pivot(values='Book-Rating', index='ISBN', columns='Age-Group')
```

```python
# train set으로 만든 ISBN별(도서별) 평점 평균
mean_rating_per_ISBN
```

```python
def get_knn_mean_rating(ISBN, k):
    
    knn_isbn_lst = get_knn_ISBNs(ISBN, k)
    
    rating_lst = []
    for ISBN, cosine in knn_isbn_lst:
        if ISBN in mean_rating_per_ISBN:
            rating = mean_rating_per_ISBN[ISBN] * cosine # cosine similarity에 따라 가중평균
        else:
            rating = total_mean_rating
        rating_lst.append(rating)
    rating_array = np.array(rating_lst)
    
    return rating_array.mean()
```

```python
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))

def baseline(user_id, ISBN):
    return total_mean_rating

def score(model):
    id_pairs = zip(test['User-ID'], test['ISBN'])
    y_pred = np.array([model(user_id, book) for (user_id, book) in id_pairs])
    y_true = np.array(test['Book-Rating'])
    return RMSE(y_true, y_pred)

def knn_score(model, k):
    id_pairs = zip(test['User-ID'], test['ISBN'])
    y_pred = np.array([model(user_id, book, k) for (user_id, book) in id_pairs])
    y_true = np.array(test['Book-Rating'])
    return RMSE(y_true, y_pred)

print(f"Baseline Model Score: {score(baseline)}")
```

```python
def ISBN_similarity_model(user_id, ISBN, k=10):
    
    return get_knn_mean_rating(ISBN, k)
```

```python
def ISBN_similarity_age_model(user_id, ISBN, k=10):
    
    knn_isbn_lst = get_knn_ISBNs(ISBN, k)
    
    rating_lst = []
    for ISBN, cosine in knn_isbn_lst:
        if ISBN in mean_rating_per_ISBN: # train set에 해당 ISBN이 있는 경우
            rating = mean_rating_per_ISBN[ISBN] * cosine # cosine similarity에 따라 가중평균
        else: # train set에 해당 ISBN이 없는 경우
            user_age = users.loc[user_id]['Age-Group'] # user_id의 age_group 정보 확인
            rating = mean_rating_per_AgeGroup[user_age] # 해당 age_group의 전체 평점 평균 반환
        rating_lst.append(rating)
    rating_array = np.array(rating_lst)
    
    return rating_array.mean()
```

```python
print(f"ISBN_similarity Model Score: {score(ISBN_similarity_model)}")
```

```python
print(f"ISBN_similarity_age_model Score: {score(ISBN_similarity_age_model)}")
```
